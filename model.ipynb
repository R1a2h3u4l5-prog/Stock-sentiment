```python
import pandas as pd
from transformers import pipeline
from collections import Counter
```

```python
df = pd.read_csv('/content/drive/MyDrive/MoneyControl Data Scraping/news.csv')

news_arr = df['News'].to_list()
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline,AutoTokenizer
import time
import warnings
import torch
import torch.nn.functional as f
warnings.filterwarnings('ignore')


model = BertForSequenceClassification.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis",num_labels=2,ignore_mismatched_sizes=True)
tokenizer = BertTokenizer.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis")


batch = tokenizer(text=news_arr,padding=True,truncation=True,max_length=512,return_tensors="pt")


with torch.no_grad():
  model_1_outputs = model(**batch)
  model_1_predictions = f.softmax(model_1_outputs.logits,dim=1)
  model_1_labels = torch.argmax(model_1_predictions,dim=1)
  model_1_original_val = [model.config.id2label[labelid] for labelid in model_1_labels.tolist()]
```

```python
print(model_1_original_val)
print(Counter(model_1_original_val))
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline,AutoTokenizer
import time
import warnings
import torch
import torch.nn.functional as f
warnings.filterwarnings('ignore')


model_2 = BertForSequenceClassification.from_pretrained("Sigma/financial-sentiment-analysis",num_labels=2,ignore_mismatched_sizes=True)
tokenizer_2 = BertTokenizer.from_pretrained("Sigma/financial-sentiment-analysis")


batch_2 = tokenizer_2(text=news_arr,padding=True,truncation=True,max_length=512,return_tensors="pt")


with torch.no_grad():

  model_2_outputs = model_2(**batch_2)
  model_2_predictions = f.softmax(model_2_outputs.logits,dim=1)
  model_2_labels = torch.argmax(model_2_predictions,dim=1)
  model_2_original_val = [model_2.config.id2label[labelid] for labelid in model_2_labels.tolist()]



```

```python
print(model_2_original_val)
print(Counter(model_2_original_val))
```

```python
df_ = pd.DataFrame({
    'News' : news_arr,
    'ahmedrachid_label' : model_1_original_val,
    'ahmedrachid_score' : [data.max() for data in model_1_predictions],
    'sigma_label' : model_2_original_val,
    'sigma_score' : [data.max() for data in model_2_predictions]
})
df_.to_csv('two_class_model_results.csv')
```
